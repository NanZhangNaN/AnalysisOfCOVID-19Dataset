# Section: 1 - Business Understanding
    The COVID-19 dataset, which is maintained by *Our World in Data*, is a collection of COVID-19 data includes data on confirmed cases, deaths, testing, ages, geographical locations, and other variables about COVID-19 or potential influencing factors about COVID-19. The data of confirmed cases and deaths in this dataset comes from the European Centre for Disease Prevention and Control (ECDC), and the data of testing for COVID-19 comes from official reports and is collected by the *Our World in Data* team, and the data of other variables are collected from other sources like United Nations, World Bank, Global Burden of Disease, Blavatnik School of Government, etc. Those organizations collect the data through a variety of sources like existing influenza and viral respiratory disease surveillance, syndromic surveillance, case reporting, commercial lab reporting, the healthcare safety system, ongoing research platforms, etc. Each row of the dataset represents the COVID-19 data of one day for a country. A collection of data makes up the COVID-19 data for a period of time in a country. And those collections of data constitute the COVID-19 data of each country over a period of time. The epidemic-related data is the basis for analysis and control of the epidemic. Through analyzing this data, users can visually see the daily or weekly patient growth rates, the cure rate, and some other related data, and so that they can get a quick understanding of the overall trends of the problems they might care. The hospitals and doctors might use this data to understand the characteristics of the infection of epidemics. And governments might use this data to get convincing feedbacks on the effectiveness of their epidemic control strategies so they can make adjustments to their epidemic control strategies or keep implementing the effective one.

## Section: 2.3 - Visualize the most important or interesting attributes using appropriate techniques.
I think new_cases_per_million and new_deaths_per_million are the most important attributes because those two variables reflect the growth trend and severity of the epidemic. I will give a brief interpretation of the graphs below in the following paragraphs.

In the graphs of New cases per million in the UK/US/World, the number of new cases ballooned between March to April, which shows us that COVID-19 outbroke began around March in the UK and US. The new cases of the UK peaked during April and May. Then it rapidly decreased until July and ballooned again in September. Through the trend, we can predict that the new cases of the UK will keep increasing in the future until reaching another peak. The new cases of the US also peaked in April and May and then became steady until July. And then the new cases of the US reached a higher peak during the mid-July and August.  Through the trend, we can predict that the new cases of the US will decrease in the future. The new cases of the World were overall upward, and we can predict that the number will keep increasing in the future.

The graphs of New deaths per million in the UK/US/World show us the number of new deaths ballooned between March to April, which corresponding to the ballooning of the number of new cases. This shows that Covid-19 is especially deadly, especially in the early stage of the outbreak. From the graphs of New deaths per million in the UK/US, we can see that the number of new deaths peaked during April and May, and then overall decreases even when the number of new cases was increasing. This trend suggests that people in the UK/US realized the severity of the epidemic and began to control it and have a better way to treats the patients who have COVID-19. From the graphs of New deaths per million in the World, we can see that the new deaths ballooned between March to April and then became steady. But it doesn’t mean the worldwide treatment has not improved because we should notice that the worldwide new cases were increasing at the same time.
***

```python
attributeName = np.array(df.columns)
#attribute = np.array(df)
worldAttr = np.array(df[df['location'].isin(['World'])])
ukAttr = np.array(df[df['location'].isin(['United Kingdom'])])
usAttr = np.array(df[df['location'].isin(['United States'])])

index_new_cases_per_million = np.argwhere(attributeName == 'new_cases_per_million')
index_new_deaths_per_million = np.argwhere(attributeName == 'new_deaths_per_million')
index_date = np.argwhere(attributeName == 'date')

uk_new_cases_per_million = ukAttr[:,index_new_cases_per_million[0,0]]
uk_new_deaths_per_million = ukAttr[:,index_new_deaths_per_million[0,0]]
uk_date = ukAttr[:,index_date[0,0]]

us_new_cases_per_million = usAttr[:,index_new_cases_per_million[0,0]]
us_new_deaths_per_million = usAttr[:,index_new_deaths_per_million[0,0]]
us_date = usAttr[:,index_date[0,0]]

world_new_cases_per_million = worldAttr[:,index_new_cases_per_million[0,0]]
world_new_deaths_per_million = worldAttr[:,index_new_deaths_per_million[0,0]]
world_date = worldAttr[:,index_date[0,0]]


l1=plt.plot(uk_date,uk_new_cases_per_million,'r',label='UK')
plt.title('New cases per million in the UK')
plt.xlabel('Date')
plt.ylabel('New cases per million')
plt.legend()
plt.show()

l4=plt.plot(uk_date,uk_new_deaths_per_million,'r',label='UK')
plt.title('New deaths per million in the UK')
plt.xlabel('Date')
plt.ylabel('New deaths per million')
plt.legend()
plt.show()

l2=plt.plot(us_date,us_new_cases_per_million,'g',label='US')
plt.title('New cases per million in the US')
plt.xlabel('Date')
plt.ylabel('New cases per million')
plt.legend()
plt.show()

l5=plt.plot(us_date,us_new_deaths_per_million,'g',label='US')
plt.title('New deaths per million in the US')
plt.xlabel('Date')
plt.ylabel('New deaths per million')
plt.legend()
plt.show()

l3=plt.plot(world_date,world_new_cases_per_million,'b',label='World')
plt.title('New cases per million in the World')
plt.xlabel('Date')
plt.ylabel('New cases per million')
plt.legend()
plt.show()

l6=plt.plot(world_date,world_new_deaths_per_million,'b',label='World')
plt.title('New deaths per million the World')
plt.xlabel('Date')
plt.ylabel('New deaths per million')
plt.legend()
plt.show()

```

And I will still take the data of the UK/US/World for example and I refer to the graphs of New cases per million in the UK/US/World drawn in section 2.3. 

Let’s recall the situation of the New cases per million graphs. From the graphs of New cases per million in the US/UK/World, we can see the number of new cases ballooned between March to April. And then the US data peaked in April and May and then became steady until July. And then the new cases of the US ballooned again and reached a higher peak during the mid-July and August. From the graph of the UK, we see that the number of new cases peaked during April and May, and then it rapidly decreased until July and ballooned again in September. In the graph of the World, the number of new cases is overall upward since March.

So we can conclude that NEW_CASE_TREND reflect the trend of total_cases_per_million or the trend of total COVID-19 cases.

## Section: 3.2 - Did you decide to implement any attribute transformations? If so, why?
I‘d like to transform the attribute population to log(population + 1). The distribution of population is concentrated in a small range, which is not convenient for users to generate related graphs or analyze the relationship between population and other variables. After transforming population to log(population + 1), the distribution becomes much more smooth, which is convenient for users to deal with it. And since the values of population contain zero, I apply log(population + 1) instead of log(population). The following graphs are the distribution of the attribute before transformation and after transformation. We can see that the second graph, which is the distribution of the log(population + 1), is much more uniform than the first one.
***

```python
df_temp = df.loc[:,["location", "population"]]

#drop null values
df_temp.dropna(axis=0, how='any', inplace=True)

arr_temp = np.array(df_temp.groupby("location")["population"].mean())
arr_temp = arr_temp[arr_temp.argsort()]

x = np.arange(-106,105)

l8=plt.plot(x,arr_temp,'r',label='population')
plt.title('population distribution')
plt.xlabel('')
plt.ylabel('populationn')
plt.legend()
plt.show()

#z-score
#df.population = (df.population-df.population.mean())/(df.population.std())

#log(x+1)
df["population"] = df["population"].apply(np.log1p)
df = df.rename(columns={'population':'log_population'})

df_temp = df.loc[:,["location", "log_population"]]
df_temp.dropna(axis=0, how='any', inplace=True)

arr_temp = np.array(df_temp.groupby("location")["log_population"].mean())
arr_temp = arr_temp[arr_temp.argsort()]

l9=plt.plot(x,arr_temp,'r',label='log_population')
plt.title('log_population distribution')
plt.xlabel('')
plt.ylabel('log_populationn')
plt.legend()
plt.show()

del df_temp
```
